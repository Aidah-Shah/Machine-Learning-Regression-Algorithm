```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.read_csv('classified-data.csv')
```

```python
data.head()
```

```python
data.describe()
```

```python
data
```

```python
data['Income (USD)'] = data['Income (USD)'].fillna(2630)
```

```python
data
```

```python
data.describe()
```

```python
data['Current Loan Expenses (USD)'] = data['Current Loan Expenses (USD)'].fillna(400.936876)
```

```python
data.describe()
```

```python
data['Dependents'] = data['Dependents'].fillna(2.000000)
```

```python
data['Credit Score'] = data['Credit Score'].fillna(739.885381)
```

```python
data['Property Age'] = data['Property Age'].fillna(2631)
data['Loan Sanction Amount (USD)'] = data['Loan Sanction Amount (USD)'].fillna(0.000000)
```

```python
data.describe()
```

```python
data['Gender'] = data['Gender'].fillna('M')
```

```python
data[pd.isnull(data['Income Stability'])]
```

```python
data['Income Stability'] = data['Income Stability'].fillna('Low')
```

```python
data.describe()
```

**FOR NaN INCOME VALUES - MEAN OF ALL INCOMES WAS USED**


**FOR NaN Current Loan Expenses (USD)	 - MEAN OF ALL Current Loan Expenses (USD) WAS USED**


**FOR NaN DEPENDENTS , CREDIT SCORE, PROPERTY AGE - MEAN OF ALL VALUES WAS USED**


**FOR NaN Loan Sanction Amount (USD) - LOAN SANCTIONED TO LOWEST 25% OF APPLICANTS(0)WAS USED TO AVOID SANCTIONING LOAN TO UNDESERVED CANDIDATES**


**FOR NaN INCOME STABILITY - INCOME STABILITY WAS SET TO LOW TO AVOID SANCTIONING LOAN TO UNDESERVED CANDIDATES**
